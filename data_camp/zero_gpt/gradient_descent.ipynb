```python
import pandas as pd
import matplotlib.pyplot as plt
weather_df = pd.read_csv('./data/clean_weather.csv', index_col=0)
weather_df = weather_df.ffill()
```

```python
weather_df.head()
```

```python
weather_df.shape
```

```python
weather_df.plot.scatter('tmax', 'tmax_tomorrow')
plt.plot([30,120], [30,120], 'green')
```

We need to use gradient descent to derive the w for our equation
$\hat_{y}$ = w_{1} * x_{1} + b

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(weather_df[['tmax']], weather_df['tmax_tomorrow'])
```

```python
weather_df.plot.scatter('tmax', 'tmax_tomorrow')
plt.plot(weather_df[['tmax']], model.predict(weather_df[['tmax']]), 'green')
```

```python
print(f'The weight is {model.coef_}')
print(f'The bias is {model.intercept_}')
```
