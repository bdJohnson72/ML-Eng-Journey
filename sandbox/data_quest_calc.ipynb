## Linear Functions
All linear functions can be written as y = mx+b
we will use the linear functions y = 3x + 1

```python
from ipywidgets import interact, interactive, Dropdown, FloatSlider, Button, Output

@interact(m=(-5, 5, 0.1), b=(-10, 10, 0.1))
def plot_linear(m=3, b=1):
    lin_xs = np.linspace(-5, 5, 100)
    lin_ys = m * lin_xs + b
    plt.figure(figsize=(8, 6))
    plt.plot(lin_xs, lin_ys, linewidth=2)
    plt.grid(True, alpha=0.3)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'Linear Function: y = {m}x + {b}')
    plt.axhline(y=0, color='k', linewidth=0.5)
    plt.axvline(x=0, color='k', linewidth=0.5)
    plt.show()
```

In a linear function m is the slope and b is the y intercept
$$m = \frac{\Delta y}{\Delta x} = \frac{y_2 - y_1}{x_2 - x_1}$$


```python
def slope(x1, y1, x2, y2):
    return (y2-y1) / (x2-x1)
slope_two = slope(5, 16, -1, -2)
slope_one = slope(0,1,4,13)
```

```python
def draw_secant(x_values):
    x = np.linspace(-20,30,100)
    y = -1*(x**2) + x*3 - 1
    plt.plot(x,y)
    # Add your code here.
    # determine y values for x
    # get slope for points
    # calc y intercept
    #plot secant line in green
    ys = []
    for val in x_values:
        y = -1*(val**2)+ val*3 -1
        ys.append(y)
    curr_slope = slope(x_values[0], ys[0], x_values[1], ys[1])
    b = ys[0] - curr_slope * x_values[0]
    secant = curr_slope * x + b
    print(curr_slope)
    print(ys)
    plt.plot(x, secant, 'g')


    plt.show()

draw_secant([3,5])
```
